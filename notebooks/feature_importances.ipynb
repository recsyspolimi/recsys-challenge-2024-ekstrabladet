```python
import joblib
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

model = joblib.load('/mnt/ebs_volume/models/catboost_new_with_recsys_train_mv_params/model.joblib')
model.get_param('cat_features')
```

```python
feature_importances = model.feature_importances_
feature_names = np.array(model.feature_names_)

sorted_importances = np.argsort(feature_importances)[::-1]
##output_path = '/mnt/ebs_volume/models/catboost_new_with_recsys_train_mv_params/feature_importances.png'

plt.figure(figsize=(10, 40))
sns.barplot(x=feature_importances[sorted_importances], y=feature_names[sorted_importances])
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.title('Catboost Feature Importances')
#plt.savefig(output_path, bbox_inches='tight')
plt.show()
```
